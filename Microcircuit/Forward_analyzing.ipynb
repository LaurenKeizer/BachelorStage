```python
import os, sys, inspect

import numpy as np

current_dir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parent_dir = os.path.dirname(current_dir)
sys.path.insert(0, parent_dir)

from brian2 import *
import Documentation.parameters as p

from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd

import Backward as forw

defaultclock = p.dt*ms
```

```python
scales  = [17,18,19,20,21,22,23]
weights = [5, 10, 15, 20, 25 ,30]
```

```python
data_PC = pd.DataFrame()
data_IN = pd.DataFrame()

for scale in scales:
    output_PC, output_IN, spiketrain = forw.forward_run(scale, weights, p.seed)
    output_PC.replace(r'\\n', ' ', regex= True)
    output_IN.replace(r'\\n', ' ', regex= True)

    data_IN = pd.concat([data_IN, output_IN])
    data_PC = pd.concat([data_PC, output_PC])
```

```python
subset = ['weights', 'scales', 'MI']
subset2 = ['L_i', 'L']
```

```python
my_data = data_PC[subset]
```

```python
x = data_PC['weights']
y = data_PC['scales']
z = data_PC['MI']

fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot(111)
scatter = ax.scatter(x, y,
                     linewidths=1, alpha=.7,
                     edgecolor='k',
                     s = 200,
                     c=z)

ax.set_xlabel('Inhibition')
ax.set_ylabel('scale')
ax.set_title('MI Feed Forward model')
pos = ax.get_position()
ax.set_position([pos.x0, pos.y0, pos.width * 0.9, pos.height])

ax.legend(*scatter.legend_elements(num=10), loc='center right', bbox_to_anchor=(1.25, 0.5), title="MI")

plt.show()
```

```python
x = data_PC['weights']
y = data_PC['scales']
F_I = data_PC['F_I']
F = data_PC['F']

fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot(111)
scatter = ax.scatter(x, y,
                     linewidths=1, alpha=.7,
                     edgecolor='k',
                     s = 200,
                     c=F_I)

ax.set_xlabel('Inhibition')
ax.set_ylabel('scale')
ax.set_title('FI_I Feed Forward model FI')
pos = ax.get_position()
ax.set_position([pos.x0, pos.y0, pos.width * 0.9, pos.height])

ax.legend(*scatter.legend_elements(num=10), loc='center right', bbox_to_anchor=(1.25, 0.5), title="FI")

plt.show()
```
